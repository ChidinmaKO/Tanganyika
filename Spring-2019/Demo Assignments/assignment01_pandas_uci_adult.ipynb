```python
import numpy as np
import pandas as pd
pd.set_option('display.max.columns', 100)
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
# we don't like warnings
# you can comment the following 2 lines if you'd like to
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../data/adult.data.csv')
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
data['sex'].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python
round(data[data['sex'] == 'Female']['age'].mean(), 2)
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
germans = data['native-country'] == 'Germany'
print('Percentage of German Citizens: {}%'.format(round(100 * germans.mean(), 2)))
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year?**

```python
high_earners = data[data['salary'] == '>50K']['age']
low_earners = data[data['salary'] == '<=50K']['age']

print('The mean age for High Earners: ',round(high_earners.mean(), 2))
print('The standard deviation for High Earners: ',round(high_earners.std(), 2))
print('\n')
print('The mean age for Low Earners: ',round(low_earners.mean(), 2))
print('The standard deviation for Low Earners: ',round(low_earners.std(), 2))
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
data['marital-status'].unique()
```

```python
data[(data['sex'] == 'Male')]['marital-status'].value_counts()
```

```python
married = data[(data['sex'] == 'Male') & data['marital-status'].str.startswith('Married')]['salary'].value_counts()
single = data[(data['sex'] == 'Male') & data['marital-status'].isin(['Never-married', 
                                                                         'Divorced', 'Separated',
                                                                        'Widowed'])]['salary'].value_counts()

print('Married Men: \n', married)
print('\n')
print('Single Men: \n', single)
# married men earn over 50k
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
data.head(3)
```

```python
long_hours = data['hours-per-week'].max()
# 99 hours!
```

```python
long_workers = data[data['hours-per-week'] == long_hours].shape[0]
# 85 people work 99 hours per week
```

```python
data[data['hours-per-week'] == long_hours]['salary'].value_counts()
# Just 25 people!
```

```python
rich_workers = data[(data['hours-per-week'] == long_hours) & (data['salary'] == '>50K')].shape[0]

print('Percentage of the people who work these long hours and receive more than 50k: {}%'
      .format(round(100 * (rich_workers/long_workers), 2)))
```
